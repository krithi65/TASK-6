#import data set

```python
import pandas as pd
df=pd.read_csv("/content/HousingData.csv")
df.head()
```

DATA CLEANING

```python
df.isnull().sum()
```

```python
# filling a null values using fillna()
df["CRIM"]=df["CRIM"].fillna(df["CRIM"].mean())
df["ZN"]=df["ZN"].fillna(df["ZN"].mean())
df["INDUS"]=df["INDUS"].fillna(df["INDUS"].mean())
df["CHAS"]=df["CHAS"].fillna(df["CHAS"].mean())
df["AGE"]=df["AGE"].fillna(df["AGE"].mean())
df["LSTAT"]=df["LSTAT"].fillna(df["LSTAT"].mean())
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.value_counts()
```

SPLIT THE DATASET

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:, :13]
Y = df["MEDV"]
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.30, random_state=10)
```

```python
x_train
```

DATA ANALYSIS

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Find correlation between all available columns
corr = df.corr()
# Find correlated data columns using heatmaps
sns.heatmap(corr)
plt.show()

# Columns which are having correlation +ve and above 0.9 are highly correlated to each other
# We can drop any of them since it will not impact on output
```

fit LINEAR REGRESSION model

```python
from sklearn.linear_model import LinearRegression
# model instantiation
linear_model = LinearRegression()

# fit the model with input and corresponding output data
linear_model.fit(x_train, y_train)
```

predict output and plot

```python
y_pred= linear_model.predict(x_test)
```

```python
y_pred
```

```python
plt.scatter(y_test, y_pred, color='pink')
plt.show()
```
